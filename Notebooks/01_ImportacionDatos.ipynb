```python
# Cargar el fichero CSV
ruta_csv = "../datos/brutos/idealista.csv"
df_idealista = pd.read_csv(ruta_csv)
```

```python
# Renombrar las columnas
df_idealista.columns = ['precio', 'distrito']

# Eliminar el primer registro
df_idealista = df_idealista.iloc[1:].reset_index(drop=True)
```

```python
df_idealista['precio'] = df_idealista.precio.str.split(expand=True)[0].str.replace('.', '').astype(int)
```

```python
# Añadimos un nuevo registro con el valor "Vicálvaro" en la columna distrito y en la columna precio el mismo valor que tenga "Moratalaz"
nuevo_registro = df_idealista[df_idealista['distrito'] == 'Moratalaz'].copy()
nuevo_registro['distrito'] = 'Vicálvaro'
df_idealista = pd.concat([df_idealista, nuevo_registro], ignore_index=True)
```

```python
# Adaptar los nombres de los distritos en df_idealista para que coincidan con listings
mapeo_distritos = {
    'San Blas': 'San Blas - Canillejas',
    'Fuencarral': 'Fuencarral - El Pardo',
    'Moncloa': 'Moncloa - Aravaca',
    # Vicálvaro ya coincide
}
df_idealista['distrito'] = df_idealista['distrito'].replace(mapeo_distritos)
```

```python
import sqlite3
from sqlite3 import Connection

# Crear la base de datos SQLite
conn: Connection = sqlite3.connect('../datos/intermedios/mercado_inmobiliario.db')

# Guardar listings
listings.to_sql('listings', conn, if_exists='replace', index=False)

# Guardar listings_det
listings_det.to_sql('listings_det', conn, if_exists='replace', index=False)

# Guardar df_idealista
df_idealista.to_sql('idealista', conn, if_exists='replace', index=False)
```

```python
# Comprobar que se han guardado correctamente
print(conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall())
print('Filas en listings:', conn.execute('SELECT COUNT(*) FROM listings').fetchone()[0])
print('Filas en listings_det:', conn.execute('SELECT COUNT(*) FROM listings_det').fetchone()[0])
print('Filas en idealista:', conn.execute('SELECT COUNT(*) FROM idealista').fetchone()[0])

conn.close()
```
